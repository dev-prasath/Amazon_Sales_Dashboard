Combining Data sets and Catalog


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
sns.set_style("whitegrid")
```

Load Data

```python
transactions = pd.read_csv("../../Data/processed_datasets/amazon_full_cleaned.csv")
products = pd.read_csv("../../Data/external/amazon_india_products_catalog.csv")

print(transactions.shape)
print(products.shape)

transactions.head()
```

Validate

```python
transactions.info()
```

```python
transactions.isnull().sum()
```

```python
transactions.duplicated().sum()
```

Merge

```python
df = transactions.merge(products, on='product_id', how='left')

print("Missing products after merge:", df['product_name_y'].isnull().sum())
```

Adding Features

```python
df['year']=pd.to_datetime(df['order_date']).dt.year
df['month']=pd.to_datetime(df['order_date']).dt.month

df['revenue']=df['discounted_price_inr'] + df['delivery_charges']
```
