```python
# Creating a new column 'payment_group' to group the payment methods into broader categories. This will help us analyze the data based on the type of payment method used by the customers.
payment_group_map = {
    'COD': 'Cash',
    'UPI': 'Digital',
    'Credit Card': 'Digital',
    'Debit Card': 'Digital',
    'Net Banking': 'Digital',
    'Wallet': 'Digital',
    'BNPL': 'Credit'
}

df['payment_group'] = df['payment_method'].map(payment_group_map)
```
